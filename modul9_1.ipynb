```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_genres = pd.read_csv('tmdb_genres.csv')
df_movies = pd.read_csv('tmdb_movies.csv')
```

```python
df_genres.head()
```

```python
df_movies.head()
```

```python
#1
q3 = np.percentile(df_movies['vote_count'], 75)

popular_movies = df_movies[df_movies['vote_count'] > q3]

popular_movies = popular_movies.sort_values(by='vote_average', ascending=False)

top_10_movies = popular_movies.head(10)
top_10_movies
```

```python
#2
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'])

date_2010 = pd.to_datetime('2010-01-01')
date_2016 = pd.to_datetime('2016-12-31')

filtered_df = df_movies[(df_movies['release_date'] >= date_2010) & (df_movies['release_date'] <= date_2016)]

grouped_df = filtered_df.set_index('release_date')[['revenue', 'budget']].resample('Y').mean()
grouped_df.index = grouped_df.index.strftime('%Y')


def million(x, pos):
        return '{:.1f}M'.format(x * 1e-6)


years = grouped_df.index
revenues = grouped_df['revenue']
budgets = grouped_df['budget']

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

formatter = plt.FuncFormatter(million)
axes.yaxis.set_major_formatter(formatter)

axes.plot(years, budgets, label='budget', color='r', linewidth=3)
axes.bar(years, revenues, label='revenue', color='royalblue')
axes.legend(loc=(1.05,0.85))
axes.set_title('Średni przychód i budżet filmu w latach 2010-2016')
```

```python
#3
df_genres.columns = ['genre_id', 'genres']
merged_data = pd.merge(df_genres, df_movies, on='genre_id')
merged_data
```

```python
#4
genres_counts = merged_data['genres'].value_counts()
most_common_genre = genres_counts.idxmax()
count_most_common_genre = genres_counts.max()

print(f"Najczęściej występujący gatunek filmu: {most_common_genre}")
print(f"Liczba filmów tego gatunku w bazie: {count_most_common_genre}")
```

```python
#5
genre_runtimes = merged_data.groupby('genres')['runtime'].mean()
most_runtime_genre = genre_runtimes.idxmax()

print(f"Gatunek filmu o najdłuższym średnim czasie trwania to: {most_runtime_genre}")
```

```python
#6
merged_data['genres'].fillna('brak danych', inplace=True)
genre_movies = merged_data[merged_data['genres'].str.contains(most_runtime_genre)]

plt.figure()
ax = genre_movies['runtime'].plot.hist(bins=20, color='skyblue', edgecolor='black')
ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
plt.xlabel('Czas Trwania (minuty)')
plt.ylabel('Liczba Filmów')
plt.title(f'Histogram czasu trwania filmów z gatunku {most_runtime_genre}')
plt.show()
```
